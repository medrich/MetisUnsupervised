```python
# imports

import os
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, WebDriverException
from bs4 import BeautifulSoup as bs
import re
import requests
import time, random
import pandas as pd
import io
from PyPDF2 import PdfFileReader
```

```python
# initialize a web driver and set the download destination

download_dir = "/home/mattiboi/Desktop/Data Science/Metis/Unsupervised Learning/project/glasgow_documents/"

options = Options()
#prefs = {"download.default_directory":"/home/mattiboi/Desktop/Data Science/Metis/Unsupervised Learning/project"}
options.add_experimental_option('prefs',  {
    "download.default_directory": download_dir,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "plugins.always_open_pdf_externally": True})

chromedriver = '/usr/bin/chromedriver'
os.environ["webdriver.chrome.driver"] = \
    chromedriver
driver = webdriver.Chrome(chromedriver, options=options)
```

```python
# create a list of pages to scrape

un_base = 'https://unfccc.int/documents?f%5B0%5D=conference%3A4301&f%5B1%5D=language%3AEnglish&search2=&search3=&page=%2C%2C'
page_counter = 0
un_urls = []

for i in range(43):
        url = un_base + str(page_counter)
        un_urls.append(url)
        page_counter += 1
```

```python
# created with assistance from Joyce Tagal

def glasgow_scraper(pages):
    root_url = 'https://unfccc.int'
    driver = webdriver.Chrome(chromedriver, options=options)
    
    for page in pages:
        # connect to the url
        driver.get(page)
        
        # read all html at the target page
        soup = bs(driver.page_source)
        
        # create an empty set to collect pdf links
        page_links = set()
        
        # find the unordered list containing links to pdfs on page
        locate = soup.find_all('ul',{'class':'list nopadding'})
        
        #find the unordered list within the main body
        for tag in locate:
            page_docs = tag.find_all('a')
            for link in page_docs:
                page_links.add(root_url + link.get('href'))
                
        print(f'Preparing to scrape this page: {page}')
                
        # downloading the pdfs on a single page to the data directory in my relevant project folder
        for url in page_links:           
            # initialize a second driver
            driver2 = webdriver.Chrome(chromedriver, options=options)
    
            # connect to the specific document page
            print(f'Now getting pdf of this document: {url}')
            driver2.get(url)
            time.sleep(1+10*random.random())
            element = driver2.find_element(By.LINK_TEXT, 'English')
            time.sleep(15)
            element.click()
            time.sleep(15)

    
    print('Scraping complete. Check your folder, there should be 422 documents.')
    driver2.close()
    driver.close()
```
